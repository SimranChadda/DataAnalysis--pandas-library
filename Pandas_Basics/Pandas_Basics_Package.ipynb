 Important Data Formatting Methods (merge, sort, reset_index, fillna)¶

```python
data2
```

```python
zoo_eats1= pd.DataFrame({'ANIMAL':['elephant','tiger','zebra','lion','kangaroo','giraffe'],'FOOD':['vegetable','meat','vegetables','meat','vegetable','vegetables']
                        })
zoo_eats1
```

```python
data2.merge(zoo_eats1)
# it will add the new column i.e food in the older data set as per the syntax given
#To keep only rows that match from the data frames, specify the argument how=‘inner’.
#By default it takes inner join method to merge the two dataset.

#inner= gives intersection of two datasets
```

```python
res= data2.merge(zoo_eats1, how='outer')
res
#To keep all rows from both data frames, specify how=‘outer’

#outer= gives union of two datasets
```

```python
res.fillna('not known')
```

```python
data2.merge(zoo_eats1,how='left')
#To include all the rows of your data frame x and only those from y that match, specify how=‘left’
```

```python
res['ANIMAL'].unique() 
#it will show all the unique values or data of a particular you want
#Similarly, nunique gives you no of unique elements in a dataframe
```
