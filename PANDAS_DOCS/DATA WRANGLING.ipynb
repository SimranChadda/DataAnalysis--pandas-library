```python
import pandas as pd   #Import the library
dataframe=pd.read_csv('Dataset/Titanic dataset.csv')  #Load the data
dataframe
```

# What information it has?
The dataset has 12 columns related to passenger details, which are,

PassengerId: Passenger’s unique ID

Survived: Survival status of the passengers (0 = No; 1 = Yes)

Pclass: Passenger class (1 = First; 2 = Second; 3 = Third)

Name: Passenger’s name

Sex: Sex of the Passenger

Age: Age of the Passenger

SibSp: Number of siblings/spouses aboard

ParCh: Number of parents/children aboard

Ticket: Ticket number

Fare: Passenger fare

Cabin: Cabin

Embarked: Port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

```python
#here we found that one of the column has spelling mistake so we can change the column name using below syntax.
dataframe.rename(columns={'2urvived':'Survived','Pclass':'Passenger Class'},inplace=True)
dataframe
```

```python
#This syntax will let you know the different categories in that particular column and its data types
dataframe['Sex'].unique()
```

```python
#value_counts will display all unique values with the number of times
#each value appears:
dataframe['Sex'].value_counts()
```

```python
#Here we want to know that how many womens are there in the titanic.This is a single condition we can use multiple condition to
dataframe[(dataframe['Sex']==1) & (dataframe['Age']<=45)]
#Here we want to know that how many female are having age greater than equal to 65.
#Note,here 1 represents Female and 0 represents male
```

```python
#pandas’ replace is an easy way to find and replace values. For example, we can
#replace any instance of "1" in the Sex column with "female":
dataframe['Sex'].replace([1,0],['Woman','Man'],inplace=True)
dataframe
#Here we only replacing the single column values.
```

```python
#The best way to delete a column is to use drop with the parameter axis=1 (i.e., the
#column axis):
dataframe.drop('Embarked',axis=1,inplace=True)
dataframe
#We can also use list of column names to drop mutiple columns
```

```python
import pandas as pd   #Import the library
data=pd.read_csv('Dataset/Titanic dataset.csv')  #Load the data
data.head()  #Show the first five rows
```

```python
data.drop(['zero.1','zero.2','zero.3','zero.4','zero.5','zero.5','zero.6','zero.7','zero.8','zero.9','zero.10','zero.11','zero.12','zero.13','zero.14','zero.15','zero.16','zero.17','zero','zero.18'],axis=1,inplace=True)
data
```

```python
#drop_duplicates defaults to only dropping rows that match
#perfectly across all columns. Under this condition, every row in our DataFrame, data
#frame, is actually unique.
data.drop_duplicates()
```

```python
#Similarly we can drop the duplicates in a particular column using below syntax
data.drop_duplicates(subset=['Sex'])
#It will display the first two rows and delete all the duplicates
```

```python
#We can display the last two rows if we want
data.drop_duplicates(subset=['Sex'],keep='last')
```

```python
#It is very common to have
#a DataFrame where each row is a person or an event and we want to group them
#according to some criterion and then calculate a statistic
```

```python
#Group rows by the values of the column 'Sex', calculate mean
# of each group
data.groupby('Sex').mean()
```

```python
#Notice Sex added after the groupby? That is because particular summary statistics
#are only meaningful to certain types of data. For example, while calculating the average
#age by gender makes sense, calculating the total age by gender does not. In this
#case we group the data into survived or not, then count the number of Sex (i.e.,
#passengers) in each group.
data.groupby('2urvived')['Sex'].count()
```

```python
data.groupby('Sex')['2urvived'].count()
```

```python
data.groupby(['Sex','2urvived'])['Age'].count()
```

```python
#apply is a great way to do data cleaning and wrangling
#Use apply to apply a built-in or custom function on every element in a column:
def uppercase(x):
    return x.upper()
data['Sex'].apply(uppercase)
```
